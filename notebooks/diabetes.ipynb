```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
import pickle
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('C:/Users/gaura/OneDrive/Desktop/My_Project/ML Project/Diabetes/Dataset/diabetes.csv')
data.head()
```

```python
data.describe()
```

```python
data['Glucose'] = data['Glucose'].replace(0,data['Glucose'].median())
data['BloodPressure'] = data['BloodPressure'].replace(0,data['BloodPressure'].median())
data['SkinThickness'] = data['SkinThickness'].replace(0,data['SkinThickness'].median())
data['Insulin'] = data['Insulin'].replace(0,data['Insulin'].median())
data['BMI'] = data['BMI'].replace(0,data['BMI'].median())
```

```python
data.describe()
```

```python
data.isna().sum()
```

```python
X = data.drop(['Outcome'],axis = 1)
y = data['Outcome']
```

# Train-Test-Split

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.25,random_state= 0)
```

```python
y_train
```

```python
def make_pickle(pickle_file_name, dump_file):
    with open(f"C:/Users/gaura/OneDrive/Desktop/My_Project/ML Project/Diabetes/Models/{pickle_file_name}.pkl","wb") as f:
        pickle.dump(dump_file,f)
```

```python
# create function for scaling and create scaling pickle file

def standard_scaler(X_train,X_test):
    scaling = StandardScaler()
    X_train_scaled = scaling.fit_transform(X_train)
    X_test_scaled = scaling.transform(X_test)
    
    make_pickle("standard_scaler", scaling)

    return X_train_scaled,X_test_scaled
```

```python
# create function for model 

def create_logistic_model(X_train,X_test,y_train):

    X_train_scaled,X_test_scaled = standard_scaler(X_train,X_test)

    logistic_model = LogisticRegression()
    logistic_model.fit(X_train_scaled,y_train)

    y_pred = logistic_model.predict(X_test_scaled)

    make_pickle('modelForPridiction',logistic_model)

    return y_pred
```

```python
def make_metrics(y_act,y_pred):
    print("Accuracy_Score - ",accuracy_score(y_act,y_pred),'\n')
    print("Confusion_Matrix - \n\n",confusion_matrix(y_act,y_pred))
```

```python
y_pred = create_logistic_model(X_train,X_test,y_train)
```

```python
make_metrics(y_test,y_pred)
```
